### Regresion lineal multiple

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


df = pd.read_csv('precio_casas_mult.csv')
df.head()
```

```python
#Separar los datos para el entrenamiento

X = df[['Tamaño_m2', 'Habitaciones', 'Distancia al Centro_km']]
Y = df['Precio']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Visualizar los datos de entrenamiento

# Crear la figura y el eje 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos
ax.scatter(x_train["Tamaño_m2"], x_train["Habitaciones"], y_train, c=y_train, cmap='viridis')

# Etiquetas
ax.set_xlabel("Tamaño de la Casa (m²)")
ax.set_ylabel("Número de Habitaciones")
ax.set_zlabel("Precio de la Casa")
ax.set_title("Datos de entrenamiento")

plt.show()
```

```python
#Entrenar el modelo 
modelo = LinearRegression()
modelo.fit(x_train, y_train)
```

```python
#Realizar predicciones en el conjunto de prueba
y_pred = modelo.predict(x_test)
```

```python
#evaluar el modelo

mse = mean_squared_error(y_test, y_pred)
r2  = r2_score(y_test, y_pred)

print(f"Error cuadratico medio {mse:.2f}")
print(f"Coeficiente de determinacion {r2:.2f}")
```

```python
#Interpretar los coeficientes del modelo

intercepto = modelo.intercept_
coeficientes = modelo.coef_

print(f"Intercepcion con el eje Y {intercepto}")
print(f"Coeficientes {coeficientes}")
```

```python
fig = plt.figure(figsize=(10,7))
ax  = fig.add_subplot(111, projection="3d")

ax.scatter(x_test['Tamaño_m2'], x_test['Habitaciones'], y_test, color="purple", alpha=0.6, label="datos reales")
ax.scatter(x_test['Tamaño_m2'], x_test['Habitaciones'], y_pred, color="red", alpha=0.6, label="predicciones")
ax.set_xlabel("Tamaño en m2")
ax.set_ylabel("Numero de habitaciones")
ax.set_zlabel("Precio")
ax.set_title("Datos reales vs predicciones")
ax.legend()
plt.show()
```
